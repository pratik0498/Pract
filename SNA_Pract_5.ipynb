```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edges_from([(1,2), (2,3) ,(3,4), (4,5), (1,3)])
```

```python
# Network Matrix Representation
matrix_representation = nx.to_numpy_array(G)
print("Network as a Matrix :\n", matrix_representation)
```

```python
# Network as an edge list
edge_list_representation = nx.to_edgelist(G)
print("Network as an Edge List:\n",edge_list_representation)
```

```python
# Network as a Sociogram (Network Graph)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_color='black', font_weight='bold')
plt.title("Network as a Sociogram")
plt.show()
```
